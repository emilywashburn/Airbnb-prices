```python
import os
import pandas as pd
```

```python
# Folder path containing CSV files
folder_path = 'C:/Users/ewashbur1/Desktop/Bootcamp/project 3/csvs'
```

```python
# Initialize an empty DataFrame to store all data
combined_data = pd.DataFrame()
```

```python
# Loop through the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        
        # Determine city and type of day from the file name
        city, day_type = filename.split("_")
        day_type = day_type.rstrip('.csv')

        # Load data into a DataFrame
        df = pd.read_csv(os.path.join(folder_path, filename))

        # Add city and day type information
        df['city'] = city
        df['day_type'] = day_type

        # Append data to the all_data DataFrame
        combined_data = combined_data.append(df, ignore_index=True)
```

```python
# Print the combined dataframe columns
print(combined_data.dtypes)
```

```python
output_file_path = 'C:/Users/ewashbur1/Desktop/Bootcamp/project 3/Airbnb-prices/Resources/combined_data2.csv'
combined_data.to_csv(output_file_path, index=False)
```
